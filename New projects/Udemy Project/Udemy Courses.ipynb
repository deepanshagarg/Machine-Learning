### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Loading the data

```python
data = pd.read_csv('udemy courses.csv')
```

```python
data.head()
```

### explaining the data
* Id: unique id for every course
* title: course title
* url: url of the course
* is_paid: Boolean value if the cousre is paid or not
* num_subscribers: number of subscribers
* avg_rating: average rating
* avg_rating_recent: average rating recently got
* rating: rating of the course
* num_reviews: number of reviews of the course
* is_wishlisted: True if its wishlisted
* num_published_lectures: number od published lectures in the course
* num_published_practice_tests: number of published practice tests in the course
* created: date created on
* published_time: published timeof the course
* discounted_price_amount: price of the course after discount
* discount_price_currency: The currency corresponding to the discounted price of a course
* discount_price_price_string: discpunted price of the course with currency
* price_detail_amount: The original price of the course
* price_detail_currency: the currency corresponding to the original price of a course for uniformaity
* price_detail_price_string: Price details with currency
